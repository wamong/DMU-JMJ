워드 클라우드 : 텍스트 데이터에서 가장 빈번하게 등장하는 단어를 시각화함
- 단어의 빈도수가 높을수록 더 크게 표시
- 텍스트 데이터에서 어떤 단어들이 많이 사용되는지 한눈에 파악 가능

1. 텍스트 데이터에서 단어를 추출
2. 각 단어의 빈도수를 계산
3. 빈도수에 따라 단어를 크기와 색상 등으로 시각화

워드 클라우드를 생성하려면

- pandas        (데이터 분석)

```python
f = open('AIwikipedia.txt', 'r', encoding='utf-8')
text = f.read()
print(text)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from wordcloud import STOPWORDS
from PIL import Image
import numpy as np
```

```python
STOPWORDS
```

```python
불용어 = STOPWORDS.union({'goal','used','use','input'})
maskCloud=np.array(Image.open('cloud.png'))
wcOutput = WordCloud(width=1500, height=1000,
                     stopwords = 불용어,
                     background_color='white',
                     mask=maskCloud).generate(text)
plt.axis('off')
plt.imshow(wcOutput)
plt.show
```

```python
wcOutput.to_file('AIwikipedia.png')
```
